# Project Nutrition Data Visualization on 80 cereal Products

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cereal_df = pd.read_csv("C:/Users/Admin/OneDrive - ptit.edu.vn/Desktop/cereal.csv")
cereal_df.head()
```

```python
#irrelevant fields

field = ['shelf', 'weight', 'cups', 'rating']

cereal_df_new = cereal_df.drop(field, axis = 1)

cereal_df_new.head()
```

```python
#corr()
cereal_corr = cereal_df_new.select_dtypes(include=['number']).corr()
cereal_corr
```

```python
#ones-like can build a matrix of boolean ( True, False) with the same Shape of our data

ones_corr = np.ones_like(cereal_corr, dtype = bool)
ones_corr
```

```python
# vif nó tương quan của 2 biến nên nó bị lặp lại
# chỉ cần lấy đồ thị phía triên hoặc phía dưới thôi
# np triu

mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data = cereal_corr, annot = True, mask = mask);
```

```python
# bị mất vitamins

adjusted_mask = mask[1:, :-1]
adjusted_mask
```

```python
adjust_cereal_corr = cereal_corr.iloc[1:, :-1]
fig, ax = plt.subplots(figsize = (10, 8))
sns.heatmap(data = adjust_cereal_corr, mask = adjusted_mask, annot = True, fmt = '.2f', cmap = 'Blues'
           ,vmin = -1, vmax = 1, linecolor='white', linewidths = 0.5)

yticks = [i.upper() for i in adjust_cereal_corr.index]
xticks = [i.upper() for i in adjust_cereal_corr.columns]

ax.set_yticklabels(yticks)
ax.set_xticklabels(xticks, rotation = 90)

title = "Correlation Matrix"


ax.set_title(title, loc = 'left', fontsize = 18)
# cmap là màu
# fmt là định dạng
# annot = true là hiện chỉ số
# maske là hàm bool 
# linecolor là đường viền nối màu
#vmin , vmax là giá trị thấp nhất và lớp nhất của colorbar
```
