```python

ca_df['Date'] = pd.to_datetime(ca_df['Date'])

ca_df.info()
```

```python
group_ca_df = ca_df.groupby("Country")

group_ca_df.first()
```

```python
group_ca_df.groups.keys()
```

```python
group_ca_df.get_group("India")
```

```python
plt.figure(figsize = (18, 14))

plt.subplots_adjust(hspace=0.5)

plt.subplot(311) # tng l 3 hng 1 l s ct, 1 l v tr subplot hng 1 ct 1
sns.lineplot(data = group_ca_df.get_group("India"), x = "Date", y = "Confirmed")
plt.title("India")

plt.subplot(312) # tng l 3 hng 1 l s ct, 1 l v tr subplot hng 2 ct 1
sns.lineplot(data = group_ca_df.get_group("China"), x = "Date", y = "Confirmed")
plt.title("China")

plt.subplot(313) # tng l 3 hng 1 l s ct, 1 l v tr subplot hng 3 ct 1
sns.lineplot(data = group_ca_df.get_group("India"), x = "Date", y = "Confirmed")
plt.title("US")
```
