## tôi muốn đổi 5k thành 5000 và 5,500 thành 5500

```python
df.head(50)
```

```python

df['peak_rpm'].apply(lambda x : (int(x.replace('K', '')) * 1000) if 'K' in x else int(x.replace(',', '')) if ',' in x else int(x))
```

```python
df['peak_rpm'].astype('int')
```

```python
def clean_peak_rpm(x):
    x = str(x)  # chắc chắn là chuỗi
    if 'K' in x:  # ví dụ "5K" -> 5000
        return int(x.replace('K', '')) * 1000
    if ',' in x:  # ví dụ "1,200" -> 1200
        return int(x.replace(',', ''))
    return int(x)  # bình thường

df['peak_rpm'] = df['peak_rpm'].apply(clean_peak_rpm)
# lúc này df['peak_rpm'] đã là int, hoặc
df['peak_rpm'] = df['peak_rpm'].astype(int)
```
