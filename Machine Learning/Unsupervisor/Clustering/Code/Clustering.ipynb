# Clustering (Phân cụm) Unsupervised
- mục tiêu nhóm các điểm dữ liệu giống nhau vào cùng 1 cụm mà không cần nhãn trước
- ví dụ phân nhóm khách hàng theo hành vi mua sắm mà không biết trước nhóm nào
## Ứng dụng
- phân nhóm khách hàng
- nhận diện bất thường
- gom tin tức, văn bản
- nén dữ liệu, giảm chiều dữ liệu
## các thuật toán Clustering phổ biến
- Kmeans: chia dư liệu thành K cụm, tối ưu khoảng cách tâm cụm, khi biết số cụm, dữ liệu có dạng cầu
- Hierarchical Clustering Gom cụp theo phân cấp ( cây), khi muốn xem cấu trúc phân cấp


# Code ví dụ Kmeans_Clustering


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# 1. Tạo dữ liệu mẫu
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# 2. Vẽ dữ liệu gốc
plt.scatter(X[:, 0], X[:, 1], s=30, cmap='viridis')
plt.title("Dữ liệu gốc (chưa phân cụm)")
plt.show()

# 3K-means
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# 4. Vẽ kết quả phân cụm
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("K-Means Clustering")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("D:/Project Python AI/Machine Learning/Unsupervisor/Clustering/Data/Iris.csv")
df.head()
```

```python
X = df.drop(columns=['Species'])

y = df['Species']
```

# K Means Clustering


```python
from sklearn.cluster import KMeans

# elbow-curve
ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50, n_init=1)
    kmeans.fit(X)
    
    ssd.append(kmeans.inertia_)
    
# plot the SSDs for each n_clusters
plt.figure(figsize=(20,8))
plt.plot(ssd)
```

```python
model = KMeans(n_clusters=3, n_init=1)

model.fit(X)
```

```python
print(model.labels_)

print(model.cluster_centers_)
```

```python
# Create a colormap
colormap = np.array(['red', 'lime', 'black'])
 
# Plot the Original Classifications
plt.scatter(X.PetalLengthCm, X.PetalWidthCm, c=colormap[model.labels_])

plt.title('K Mean Clustering')
```
