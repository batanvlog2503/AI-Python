# PCA (Principal Component Analysis) ( Unsupervisor)


```python
# - mục tiêu: giảm số chiều dữ liệu(dimessionality reduction) trong khi vẫn giữ lại nhiều thông tin
# - giúp giảm nhiều dữ liệu
# - trực quan hóa dữ liệu cao chiều
# - tăng tốc độ huấn luyện ML
```

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

# 1. Load dữ liệu mẫu
data = load_iris()
X = data.data
y = data.target

# 2. Chuẩn hóa dữ liệu
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Khởi tạo PCA (chọn 2 thành phần để trực quan hóa)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print("Tỷ lệ phương sai giữ lại:", pca.explained_variance_ratio_)
print("Tổng phương sai giữ lại:", sum(pca.explained_variance_ratio_))

# 4. Vẽ biểu đồ
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='viridis')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA trên Iris dataset")
plt.colorbar(label="Classes")
plt.show()
```
