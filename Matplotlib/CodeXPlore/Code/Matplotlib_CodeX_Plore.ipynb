
## Object Oriented APU

```python
#1. Prepare Data
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8,4))

ax[0].plot([1,2,3],[1,4,9])
ax[0].set_title("Biu  1")

ax[1].plot([1,2,3],[2,4,6])
ax[1].set_title("Biu  2")

plt.show()
```

```python
#OO API

rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000 * rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s = sizes, c = colors, cmap = "viridis", alpha = 0.3)

fig.colorbar(img1)
```

## 4. Histogram biu din tn sut


```python
#prepare Datas
student_height = np.random.normal(170, 10, 250)
student_height[:10]
```

```python
# Pyplot API
# bin khong nh hn
bins = np.arange(145, 191, 5) # cch 5
plt.hist(student_height, bins = bins, align = 'right', range = [160, 191])
```

## Subplots() v nhiu  th trn cng mt figure

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))

#plot data to each axis
ax1.plot(x, x / 2)
ax1.plot(x, x ** 2)

#ax2 scatter
ax2.scatter(np.random.random(10), np.random.random(10))

#ax3 bar
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values())

#ax4 hist
ax4.hist(student_height, bins = bins, align = 'right', range = [160, 191])
```
