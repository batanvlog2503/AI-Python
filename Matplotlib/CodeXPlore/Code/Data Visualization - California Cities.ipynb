```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv("C:/Users/Admin/OneDrive - ptit.edu.vn/Desktop/california_cities.csv")
```

```python
# Extract Latd vĩ độ and longd =  kind độ
plt.style.use('seaborn-v0_8-whitegrid')
lat , lon = cities['latd'], cities['longd']


population, area = cities['population_total'], cities['area_total_km2']
plt.figure(figsize = (8, 6)) # rộng 8 inch dài 4 inch
#Plot using Pyplot API

plt.scatter(lon, lat, c = np.log10(population), cmap = 'viridis', s = area,
           linewidths = 0, alpha = 0.5)
plt.axis('equal')
plt.xlabel('longlatitude')
plt.ylabel('latitude')
plt.colorbar(label = 'log$_{10}$(population)');

plt.clim(0, 10) # rút gọn thanh bar màu

# set title
plt.title("California Cities: Population and Area Distribution")

#Create A legend for cities sizes

area_range = [50, 100, 300, 500]

for area in area_range:
    plt.scatter([], [], s = area, label = str(area) + "km$^2$") # KHÔNG CÓ TỌA ĐỘ X, Y
plt.legend(labelspacing = 1, title = 'City Area')
```
